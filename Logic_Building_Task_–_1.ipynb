1. User Login Check


```python
username = "admin"
password = "1234"

input_username = input("Enter Username: ")
input_password = input("Enter Password: ")

if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Login UNsuccessful")
```

 2. Pass / Fail Analyzer


```python
marks = [45, 78, 90, 33, 60]


pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

3. Simple Data Cleaner


```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Original list:", names)
print("Cleaned list:", cleaned_names)
```


4. Message Length Analyzer


```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print("Message:", msg)
    print("Length:", length)

    if length > 10:
        print("Status: Long Message ")
    else:
        print("Status: Short Message")

    print("----------------------")
```

5. Error Message Detector


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
